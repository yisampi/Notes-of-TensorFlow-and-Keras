```python
with open('/etc/resolv.conf') as f:
    while True:
        line = next(f, None)
        if line is None:
            break
        print(line, end='')

print('@@@@@@')

with open('/etc/resolv.conf') as f:
    for l in f:
        print(l, end='')
```
